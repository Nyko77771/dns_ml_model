```python
data = pd.read_csv('balanced_combined.csv')
data.head()
```

```python
# Dropping Columns
data = data.drop(columns={'Unnamed: 0.1','Unnamed: 0'})
data.head()
```

```python
# Creating Extra Informational Columns
# Based on URL data
```

```python
# Adding URL Length
data['Length'] = 0
for index in data.index:
    data.loc[index, 'Length'] = len(data.loc[index, 'URL'])
```

```python
data.head()
```

```python
# Adding https analyses
data['Contains_HTTPS'] = 0
for index in data.index:
    if 'https' in data.loc[index, 'URL']:
        data.loc[index, 'Contains_HTTPS'] = 1
```

```python
data.head()
```

```python
# IP address presence column
data['Has_IP'] = 0
pattern = r'(\d{1,3}\.){3}\d{1,3}'
data['Has_IP'] = data['URL'].str.extract(pattern).notna().astype(int)
data.head()
```

```python
# Digit Counting
data['Digit_Count'] = 0
for index in data.index:
    count = 0
    for c in data.loc[index, 'URL']:
        if c.isdigit():
            count += 1
    if count > 0:
        data.loc[index, 'Digit_Count'] = count
```

```python
data.head()
```

```python
data['Dot_Count'] = 0
for index in data.index:
    count_dots = 0
    string = data.loc[index, 'URL']
    count_dots = string.count('.')
    if count_dots > 0:
        data.loc[index, 'Dot_Count'] = count_dots
data.head()
```

```python
# Counting Sub-domains Present
data['Has_Subdomain'] = 0
data['Subdomain_Count'] = 0
for index in data.index:
    sections = tld.extract(data.loc[index, 'URL'])
    count = len(sections.subdomain.split('.'))
    if count > 1:
        data.loc[index, 'Has_Subdomain'] = 1
        data.loc[index, 'Subdomain_Count'] = (count-1)
data.head()
```

```python
# Counting Hashes in URL
data['Hyphen_Count'] = 0
for index in data.index:
    data.loc[index, 'Hyphen_Count'] = data.loc[index, 'URL'].count('-')
```

```python
data.head()
```

```python
# Creating Column for Counting Other Special Characters
data['Special_Count'] = 0
for index in data.index:
    count = 0
    for c in data.loc[index, 'URL']:
        if not c.isalnum():
            if c != '.' and c != '/':
                count += 1
                data.loc[index, 'Special_Count'] = count
data.head()
```
