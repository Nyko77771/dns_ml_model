```python
# Importing DGA csv

dga = pd.read_csv('dga.csv')
dga.head()
```

```python
# General Information about DGA csv:
dga.info()
```

```python
# Method for checking Duplicates:
def check_duplicates(data_frame):
    print('Starting Duplicate Check')
    try:
        if data_frame.duplicated().any():
            print('Found Duplicates')
        else:
            print('No Dupliicates Found')
    except Exception as e:
        print('Exception found in check_duplicate function')
        print(f'Exception: {e}')
```

```python
# Cleaning DGA csv:

# 1. Looking for duplicates
check_duplicates(dga)
```

```python
# Checking which rows are duplicated
duplicates = dga[dga.duplicated()]
print('Found Duplicates:')
print(duplicates)
```

```python
# Removing duplicates
dga = dga.drop_duplicates()
print('Checking Duplicates Again')
check_duplicates(dga)
```

```python
# Getting Column Information function
def columns_type(data_frame):
    print('Starting column type check')
    
    try:
        category_column = [col for col in dga.columns if dga[col].dtype == 'object']
        number_column =  [col for col in dga.columns if dga[col].dtype != 'object']
        print(f'Category columns: {category_column}')
        print(f'Number columns: {number_column}')
    except Exception as e:
        print('Exception occured in columns_type function')
        print(f'Exception: {e}')
```

```python
# Checking Column Type for DGA
columns_type(dga)
```

```python
def get_nulls(data_frame):
    print('Total nulls per column:')
    null_sum = data_frame.isnull().sum()
    print(null_sum)
```

```python
get_nulls(dga)
```

```python
# Function for inserting Threat Column
def insert_threat_type(data_frame):
    try:
        print('Adding Threat Value')
        data_frame['Threat'] = 1
        print('Finished Creating Threat Column')
    except Exception as e:
        print('An exception occurred in insert_threat function')
        print(f'Exception: {e}')
```

```python
# Removing Unneccessary Columns
dga = dga.drop(columns=['Group'])
dga.head()
```

```python
# Checking and Converting Domain to URL
def check_convert_domain(data_frame):
    try:
        print('Starting Check')
        has_port = data_frame['domain'].str.startswith(('https://','http://'))
        has_www = data_frame['domain'].str.startswith(('www.' 'https://www.','http://www.'))
        if has_port.any():
            print('Found Ports')
        else:
            print('Ports Not Found')
        print('Starting Conversion')
        data_frame['URL'] = data_frame['domain']
        data_frame.loc[~has_port, 'URL'] = ('https://' + data_frame.loc[~has_port, 'URL'])
        data_frame.loc[~has_www, 'URL'] =(data_frame.loc[~has_www, 'URL'].str.replace('https://','https://www.',regex=False))
        print('Conversion Finished')
    except Exception as e:
        print('Exception occurred in check_convert_domain function')
        print(f'Exception: {e}')    
```

```python
check_convert_domain(dga)
dga.head()
```

```python
# Adding Threat Column to DGA
insert_threat_type(dga)
```
