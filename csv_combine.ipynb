### Joining

```python
os.chdir(clean_path + '/to_join/')
os.getcwd() 
```

```python
files = '*.csv'
joined_list = glob.glob(files)
combined = pd.concat(map(pd.read_csv, joined_list), ignore_index=True)
combined.head()
```

### Cleaning Combined File

```python
# Reducing Size
```

```python
# Removing N/A
total_na = combined.isna().sum()
print(total_na) # No N/A
```

```python
# Removing Duplicates
total_duplicates  = combined.duplicated().sum()
print(total_duplicates)
```

```python
combined = combined.drop_duplicates()
```

```python
combined_size = combined.shape
print(combined_size)
```

```python
if not os.path.isdir('combined'):
    os.mkdir('combined')
os.chdir('combined')
os.getcwd()
```

```python
combined.to_csv('combined.csv') # Makes a very large file still
```

```python
# Saving as gzip instead
combined.to_csv("combined.csv.gz", index=False, compression="gzip")
```

# Creating a Smaller File

```python
randomised_combined = combined.sample(frac=.3).reset_index(drop=True)
```

```python
print(randomised_combined.shape)
```

```python
randomised_combined.head()
```

```python
items = randomised_combined['Threat'].unique()
print(items)
```

```python
num_1 = (randomised_combined['Threat'] == 1).sum()
num_0 = (randomised_combined['Threat'] == 0).sum()
print(f"Number of 1's: {num_1}")
print(f"Number of 0's: {num_0}")
```

```python
# Too many malicious examples in a data sample
print(f'Ratio Malicious/Benign samples: {num_1/num_0}')
```

```python
# Spliting into Major and Minor
major = randomised_combined[randomised_combined['Threat'] == 1]
minor = randomised_combined[randomised_combined['Threat'] == 0]
```

```python
major.head()
```

```python
minor.head()
```

```python
major_size = major.shape
minor_size = minor.shape
print(major_size)
print(minor_size)
```

```python
major_normal_size = major.sample(n=len(minor))
major_size_fig = major_normal_size['Threat'].sum()
print(major_size_fig)
```

```python
balanced_combined = pd.concat([major_normal_size, minor]).sample(frac=1).reset_index(drop=True)
```

```python
balanced_num_1 = (balanced_combined['Threat'] == 1).sum()
balanced_num_0 = (balanced_combined['Threat'] == 0).sum()
print(f"Number of balanced 1's: {balanced_num_1}")
print(f"Number of balanced 0's: {balanced_num_0}")
```
