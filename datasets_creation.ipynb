```python
# Importing Majestic Million Top 1 Million Visited Domains
os.chdir(good_csv_path)
mm_data = pd.read_csv('majestic_million.csv')
mm_data.head()
```

```python
# Removing Irrelevant Columns
mm_data = mm_data.drop(columns=['GlobalRank', 'TldRank', 'TLD', 'RefSubNets', 'RefIPs', 'IDN_Domain', 'IDN_TLD', 'PrevGlobalRank', 'PrevTldRank', 'PrevRefSubNets', 'PrevRefIPs'])
mm_data.head()
```

```python
# Creating a URL Column Function
def url_column(data_frame):
    try:
        print('Converting Domain Names to URL address')
        has_protocol = data_frame['Domain'].str.startswith(
            ('http://','https://')
        )
        has_www = data_frame['Domain'].str.startswith(
            ('www.', 'http://www.', 'https://www.')
        )

        if has_protocol.any():
            print('Protocol Found')
        else:
            print('Protocol Not Found')
            
        data_frame['URL'] = data_frame['Domain']
        data_frame.loc[~has_protocol, 'URL'] = ('https://' + data_frame.loc[~has_protocol, 'Domain'])
        data_frame.loc[~has_www, 'URL'] =(data_frame.loc[~has_www, 'URL'].str.replace('https://','https://www.',regex=False))
        print('Conversion Completed')
        return data_frame
    except Exception as e:
        print('Exception occured in url_column function')
        print(f'Exception: {e}')
        return None
```

```python
# Converting Domain to URL
mm_data = url_column(mm_data)
```

## Saving

```python
### Saving good_csv function
def saving_good_csv(data_frame, name):
    try:
        print('Starting Saving')
        os.chdir(root_path)
        if name == 'Cisco':
            file_name = 'cisco.csv'
        elif name == 'Majestic':
            file_name = 'majestic.csv'
        elif name == 'Kaggle_Balanced':
            file_name = 'kaggle_balanced.csv'
        else:
            print('Name not recognised')
            return
             
        if not os.path.isdir('imported_data'):
            print('Making imported_data directory')
            os.mkdir('imported_data')
        os.chdir(root_path + '/imported_data')
        if not os.path.isdir('good_csv'):
            print('Making good_csv directory')
            os.mkdir('good_csv')
        os.chdir(root_path + '/imported_data/good_csv')
        data_frame.to_csv(root_path + '/imported_data/good_csv/' + file_name, index=False)
        print('Save Completed')
    except Exception as e:
        print('Exception occurred')
        print(f'Exception: {e}')
```
